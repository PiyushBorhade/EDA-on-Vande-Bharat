**Importing required libraries!**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Vande Bharat.csv")
df
```

**Different Operating Zones**

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Operator',data=df)
plt.title("Opearting Zones")
plt.xlabel("Zone")
plt.ylabel("Train Counts")
plt.show()
```

**Vande Bharat Express - Source/Originating Station**

```python
plt.figure(figsize=(12,6))
sns.countplot(y='Originating City',data=df)
plt.title("Staring point city wise")
plt.ylabel("Originating City")
plt.xlabel("Trains")
plt.show()
```

**Vande Bharat Express - Terminal Cities**

```python
plt.figure(figsize=(12,6))
sns.countplot(y='Terminal City',data=df)
plt.title("Terminating point city wise")
plt.ylabel("Terminal City")
plt.xlabel("Trains")
plt.show()
```

```python
df.columns
```

**Frequency of train on days**

```python
plt.figure(figsize=(12,6))
sns.countplot(y=df['Frequency'])
plt.title("Distance Travelles city wise")
plt.ylabel("Distance travelled")
plt.xlabel("Train name")
plt.show()
```

**Originating Cities - Zone wise**

```python
df.columns
```

```python
df[['Originating City','Terminal City','Operator','Speed']].groupby(['Originating City','Terminal City','Operator','Speed']).size()
```

**Speed of Vande Bharat with respect to Operator**

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df,x='Operator', y='Speed')
plt.xticks(rotation=45, ha='right')
plt.title('VB Speed (Zone wise)')
plt.show()
```

**Vande bharat speed!**

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,y='Train Name', x='Speed')
plt.xticks(rotation=45, ha='right')
plt.title('VB Speeds')
plt.show()
```

**Distance travelled by Vande Bharat**

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df,y='Train Name', x='Distance')
plt.xticks(rotation=45, ha='right')
plt.title('VB Speeds')
plt.show()
```

**Vande Bharat Express - Inauguration Year**
